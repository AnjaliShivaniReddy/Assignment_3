# 1. Enhancing Images

```python
# importing the required libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Loading the image
image = cv2.imread("chest.png", cv2.IMREAD_GRAYSCALE)

# Defining the kernel
kernel = np.array([[-1, -2, -1],
                   [-2, 13, -2],
                   [-1, -2, -1]])

# Applying convolution to perform edge enhancement
enhanced_image = cv2.filter2D(image, -1, kernel)

# Displaying the original and enhanced images
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(enhanced_image, cmap='gray')
plt.title('Edge Enhanced Image')

plt.show()

# Saving the enhanced image
cv2.imwrite("enhanced_chest.png", enhanced_image)
```

# Explaination

1. **Original Image (Left):**
   - The left subplot shows the original grayscale image "chest.png" without any enhancement.
   - This image appears as it is, with its original contrast and details.

2. **Edge Enhanced Image (Right):**
   - The right subplot displays the result of applying edge enhancement to the original image using the defined kernel.
   - Edge enhancement is a process that highlights the edges and fine details in the image.
   - The edges in the image are sharpened, making them more prominent.
   - Areas with significant changes in pixel intensity, such as the outlines of objects, will be more pronounced in this enhanced version.

3. **Saved Enhanced Image:**
   - The code also saves the enhanced image as "enhanced_chest.png" in the same directory.

The edge enhancement process aims to improve the visibility of important features in the image, making it easier to identify edges and fine structures, which can be useful in various image processing applications.
